```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# Load the data into a pandas DataFrame
df = pd.read_csv('/Users/mawulitsimese/Downloads/zscore1.csv')  # Replace 'your_data.csv' with the name of your data file


df
```

```python
df = df[df['at'].notna()]
df = df[df['lt'].notna()]
df = df[df['prcc_c'].notna()]
df = df[df['csho'].notna()]
df = df[df['re'].notna()]
df = df[df['ebit'].notna()]
df = df[df['sale'].notna()]
df = df[df['act'].notna()]
df = df[df['lct'].notna()]
df
```

```python
# Calculate Altman's Z-score
df['zscore'] = 1.2 * ((df['act'] - df['lct']) / df['at']) + 1.4 * (df['re'] / df['at']) + 3.3 * (df['ebit'] / df['at']) + \
                0.6 * ((df['prcc_c'] * df['csho'])/ df['lt']) + 1.0 * (df['sale'] / df['at'])
```

```python
# Output the results
df[['gvkey', 'datadate', 'zscore']].to_csv('zscore_results.csv', index=False)  # Replace 'zscore_results.csv' with the desired output file name
```

```python


# Drop rows with NaN values
df['zscore'].dropna(inplace=True)

# Verify if NaN values are removed
print("Number of rows with NaN values removed: ", len(df['zscore']))
```

```python

# Remove all rows with blanks in the ZSCORE column
df = df[df['zscore'].notna()]
df
```

```python
##describing the zscore
print(df['zscore'].describe())
```

```python
from scipy.stats.mstats import winsorize
import seaborn as sns

# Winsorization transformation
df['zscore_winsorized'] = winsorize(df['zscore'], limits=[0.01, 0.01])

# Summary statistics of the winsorized variable
print(df['zscore_winsorized'].describe())

# Distribution plot of the winsorized variable
sns.histplot(data=df, x="zscore_winsorized", kde=True)

# Skewness and Kurtosis of the winsorized variable
print("Skewness: ", df['zscore_winsorized'].skew())
print("Kurtosis: ", df['zscore_winsorized'].kurtosis())
```

```python
#drop NAs in the containers
df = df[df['ceq'].notna()]
df = df[df['dltt'].notna()]
df = df[df['che'].notna()]
df = df[df['ib'].notna()]
df = df[df['dlc'].notna()]
df = df[df['sale'].notna()]
df
```

```python
import numpy as np

# Calculate logarithm of firm size
df['log_at'] = np.log(df['at'])

```

```python
# Calculate Tobin's q
df['tobins_q'] = (df['ceq'] + df['dltt']) / df['at']



```

```python
# Calculate book leverage
df['book_leverage'] = df['dltt'] / df['at']

```

```python
# Calculate firm-level cash-holding
df['che_at'] = df['che'] / df['at']

```

```python
# Calculate firm's operating profitability
df['operating_profitability'] = df['ib'] / df['sale']

```

```python
# Calculate firm-level liquidity
df['current_ratio'] = df['act'] / df['dlc']

```

```python
# Create a dummy variable for dividend payment
df['dividend_dummy'] = np.where(df['dvt'] > 0, 1, 0)
```

```python
df = df[df['operating_profitability'].notna()]
df = df[df['current_ratio'].notna()]
df
```

```python
##i have inf in my dataset....
import pandas as pd

# Replace 'inf' with NaN
#df['current_ratio'].replace([float('inf'), float('-inf')], pd.NA, inplace=True)
# Replace 'inf' with NaN
df['current_ratio'][df['current_ratio'] == float('inf')] = np.nan
df['operating_profitability'][df['operating_profitability'] == float('-inf')] = np.nan
df
```

```python
#removing NAs now!!!!!!
df = df[df['current_ratio'].notna()]
df = df[df['operating_profitability'].notna()]
df
```

```python
from scipy.stats.mstats import winsorize
import pandas as pd

# Define the list of variables to be winsorized
variables_to_winsorize = ['log_at', 'tobins_q', 'book_leverage', 'che_at', 'operating_profitability', 'current_ratio', 'dividend_dummy']

# Winsorize the variables at the 1% level on both tails
for variable in variables_to_winsorize:
    df[variable] = winsorize(df[variable], limits=[0.01, 0.01])

# Display summary statistics for all the explanatory variables
summary_statistics = df[variables_to_winsorize].describe()
print(summary_statistics)
```

```python


# Create a dummy variable based on ZSCORE measure
df['zscore_measure'] = 0  # Initialize the new dummy variable with 0
df.loc[df['zscore'] < 1.8, 'zscore_measure'] = 1  # Set the value to 1 if ZSCORE is below 1.8
df.loc[df['zscore'] > 3, 'zscore_measure'] = 0  # Set the value to 0 if ZSCORE is above 3
 
```

```python
df
```

```python
from sklearn.linear_model import LogisticRegression

# Linear Regression
X_linear = df[['log_at', 'tobins_q', 'book_leverage', 'che_at','operating_profitability','current_ratio']]  
X_linear = sm.add_constant(X_linear)  # Add constant term to the model
y_linear = df['zscore_measure']
model_linear = sm.OLS(y_linear, X_linear)
results_linear = model_linear.fit()

# Print the linear regression results
print("Linear Regression Results:")
print(results_linear.summary())

# Logistic Regression
X_logit = df[['log_at', 'tobins_q', 'book_leverage', 'che_at','operating_profitability','current_ratio']] 
y_logit = df['zscore_measure']
model_logit = LogisticRegression()
model_logit.fit(X_logit, y_logit)

# Print the logistic regression coefficients
print("Logistic Regression Coefficients:")
print(model_logit.coef_)
```
